```python
data = pd.read_csv("Jeton - 01.csv")
data.head(5)
```

```python
data.info()
```

```python
data = data[['Update Date', 'Currency Code', 'Funds Amount',
               'Transaction Amount', 'Transaction Type', 'Verb',
               'Transaction Status', 'Fee', 'Balance',
               'Payment Id', 'Merchant Order Id', 'Description',
               'Jeton transaction ID', 'Account ID', 'Fx Rate',
               'Original date(Created date)']]
```

```python
# Распределение данных по датам

data['Update Date'] = pd.to_datetime(data['Update Date'])

data['Date'] = data['Update Date'].dt.date
data['Date'].value_counts()
```

```python
# Отсортируем данные по дате

data.sort_values(by=['Update Date'],inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
# Значения переменной Description

data['Description'].value_counts(dropna=False)
```

```python
# Значения переменной Transaction Type

data['Transaction Type'].value_counts(dropna=False)
```

```python
# Оставляем тллько Pay и Payout транзакции

data = data[data['Transaction Type'] != 'DECREASE_BALANCE']

data.reset_index(drop=True, inplace=True)
data['Transaction Type'].value_counts(dropna=False)
```

```python
# переменная Verb

data['Verb'].value_counts(dropna=False)
```

```python
# Преобразуем строки в числа (string -> float)

data['Transaction Amount'] = data['Transaction Amount'].apply(lambda x: x.replace(',','.')).astype(float)
data['Balance'] = data['Balance'].apply(lambda x: x.replace(',','.')).astype(float)
```

```python
c = [-3.46]

# Считаем разницу между текущей транзакцией и предыдущей
for a,b in data.iterrows():
    if a != 0:
        x = data.loc[a,'Balance'] - data.loc[a-1,'Balance']
        x = round(x ,2)
        c.append(x)
        
data['Computed_diff'] = c
```

```python
print('Совпало посчитанное значение с тем что оказалось в Transaction Amount :',
                      data[(data['Computed_diff'] == data['Transaction Amount'])].shape[0],
      '\nНе совпало посчитанное значение с тем что оказалось в Transaction Amount :', 
                      data[(data['Computed_diff'] != data['Transaction Amount'])].shape[0])
```

```python
# Основная гипотеза была в том, что на баланс влияет переменная Transaction Amount.
# Чтобы это проверить я посчитал разницу балансов для каждой транзакции руками и сравнил полученные значения Transaction Amount.
# Как видно, в 33 случаях, что-то идёт не так. Ниже я вынес 2 кейса: 
                                 # в первом это одна транзакция,
                                 # во втором серия транзакций.
# Почему так происходит я не смог уловить, аномалий,которые смогли бы это объяснить, в других переменных я не увидел.



# Разница между итоговым балансом и исходным не равна сумме транзакций. Показал ниже
```
